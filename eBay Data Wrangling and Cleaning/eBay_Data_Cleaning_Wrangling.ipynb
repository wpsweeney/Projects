<h2 style="color:skyblue;"><strong>eBay Data Wrangling</strong></h2>

```python
#load pandas library
import pandas as pd
#hide long warning messages from output
pd.options.mode.chained_assignment = None  # default='warn'

#load the data
LaptopData = pd.read_csv('EbayPcLaptopDataUnclean.csv')
PriceData = pd.read_csv('EbayPcLaptopPriceData.csv')

#preview LaptopData
LaptopData
```

```python
#preview PriceData
PriceData
```

```python
#merge the data
MergedData = pd.merge(LaptopData, PriceData, on='Item Number', how='left')

#preview MergedData
MergedData
```

```python
#explore the data types
MergedData.dtypes
```

```python
# Start by removing any potential duplicates
duplicates = MergedData.duplicated()
print(duplicates.sum())

#drop duplicates
MergedData = MergedData.drop_duplicates()
```

```python
# Confirm that the duplicates have been removed
duplicates = MergedData.duplicated()
print(duplicates.sum())
```

```python
# Check for NULL values
MergedData.isna().sum() / MergedData.__len__() * 100
```

```python
# Nearly all of the data is missing from the 'Rating' and 'Ratings Count' column, so we will drop it
MergedData = MergedData.drop(columns=['Rating', 'Ratings Count'])

# Same issue for 'Manufacturer Color' and 'Country Region of Manufacture', so we will drop them
MergedData = MergedData.drop(columns=['Manufacturer Color','Country Region Of Manufacture'])

```

<h4 style="color:skyblue;">Since many of these fields are optional for eBay listings, we can FillNA with 'Unknown' to salvage some of the data. However, we will only do this for fields that are not critical to the computer specifications impacting price.</h4>

<h5 style="color:skyblue;">I am considering the hardware specs, brand/model, and condition to be the most important factors. This is because a laptop model can vary in price by $1000+ simply based on the different hardware options, including processor, RAM, screen size, storage, and GPU. Prices also significantly vary depending on the brand or model of the laptop, even if the hardware is the same. This is mostly the result of hardware/material quality differences, brand reputation, and even proprietary software. Finally, for obvious reasons, used laptops sell for less than the same laptop when brand new.<h5>

```python
# fill NA values with 'Unknown' for the following columns
for i in ['Color','Type','Release Year','Maximum Resolution','OS','Features','Seller Note']:
    MergedData[i].fillna('Unknown', inplace=True)
```

```python
#confirm that the NAN rows have been populated with a default value
MergedData.isna().sum() / MergedData.__len__() * 100
```

<h4 style="color:skyblue;"> We used FillNA to salvage data, but we want to drop the rows that are missing critical analysis data. This leaves over 500 records, which is plenty for analysis in this case.</h4>

```python
#drop rows with missing values in the remaining columns
MergedData_Sample = MergedData.dropna()
print(MergedData_Sample.count())
```

```python
MergedData_Sample
```


<h2 style="color:skyblue;"><strong>Clean Remaining Data</strong></h2>
<p><strong>Now that we have a solid sample, we need to clean the data to make the values as consistent as possible</strong></p>


<h4 style="color:skyblue;">Starting with Ram Size<h4>

```python
MergedData_Sample['Ram Size'].unique()
```

```python
# Add a 'Multiple' option when multiple sizes are available
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['8GB / 16 GB OR 32GB SELECT FROM ABOVE', '8GB,  16GB,  32GB,  64GB','4 GB- 8 GB', 'Up to 64GB', 'Up to 32GB', 'Up to 36 GB', 'Up to 40GB', 'Up to 32 GB', '4GB - 16 GB'], 'Multiple Options')

# Standardize GB formats of the Ram Size column
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['8GB DDR4','8GB','8 GB DDR4 Installed---Can upgrade o 32GB DDR4'], '8 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['20gb'], '20 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['4GB'], '4 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['64 MB'], '64 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['12gb','12GB'], '12 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['24GB'], '24 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['16 GB RAM', '16GB','16 GB DDR4'], '16 GB')
MergedData_Sample['Ram Size'] = MergedData_Sample['Ram Size'].replace(['32GB'], '32 GB')

MergedData_Sample['Ram Size'].unique()
```

<h4 style="color:skyblue;">SSD Capacity<h4>

```python
MergedData_Sample['SSD Capacity'].unique()
```

```python
# Create two new outputs: 'Unknown' and 'Multiple Options'
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].replace(['1 Used,  1 support Max 1TB M.2 Extend','na','0','Not Include','NONE','Not Applicable','.',], 'Unknown')
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].replace(['Up to 2TB','FROM 120GB TILL 1TB SSD SELECT FROM ABOVE','128GB,  256GB,  500GB,  1TB,  2TB','Up to 1 TB','256 GB / 512GB / 1TB SSD','Up to 2 TB','120GB - 1TB','120 GB - 240 GB','120 GB - 500 GB'], 'Multiple Options')

# Trying a different approach to standardize the valid values - convert to upper case (reduce # of unique values) and then correct using a dictionary
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].str.upper()
ssd_capacity_mapping = {
    '1TB': '1 TB',
    '1 TB(NOT IN)': '1 TB',
    '64GB': '64 GB',
    '4TB': '4 TB',
    '32GB': '32 GB',
    '128GB': '128 GB',
    '500 GB HDD': '500 GB',
    '931.51GB': '931.51 GB',
    '256 GBSSD': '256 GB',
    '16GB': '16 GB',
    '16 GB SSD': '16 GB',
    '512GB': '512 GB',
    '256GB': '256 GB',
    '512': '512 GB',
    '2TB': '2 TB',
    '180GB': '180 GB',
    '256SSD': '256 GB',
    '512 GB SSD': '512 GB',
    '120GB': '120 GB',
    '240GB': '240 GB'
}

# Replace using the mapping dictionary
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].replace(ssd_capacity_mapping)

# Remove upper from plain text values
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].replace('UNKNOWN', 'Unknown')
MergedData_Sample['SSD Capacity'] = MergedData_Sample['SSD Capacity'].replace('MULTIPLE OPTIONS', 'Multiple Options')


MergedData_Sample['SSD Capacity'].unique()
```

<h4 style='color:skyblue;'>Condition</h4>

```python
MergedData_Sample['Condition'].unique()
```

```python
# If 'new' is included within the first 10 characters of the 'Condition' column, replace the value with 'New'
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'New' if 'NEW' in x[:10].upper() else x)

# The test worked, so we'll the same for other common conditions
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Used' if 'USED' in x[:10].upper() else x)
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Very Good' if 'VERY GOOD' in x[:10].upper() else x)
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Excellent' if 'EXCELLENT' in x[:10].upper() else x)
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Good' if 'GOOD' in x[:10].upper() else x)
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Open Box' if 'OPEN BOX' in x[:10].upper() else x)
MergedData_Sample['Condition'] = MergedData_Sample['Condition'].apply(lambda x: 'Certified Pre-Owned' if 'CERTIFIED' in x[:10].upper() else x)


# To remove any other uncommon conditions, drop rows in which the 'Condition' column is longer than 20 characters
MergedData_Sample = MergedData_Sample[MergedData_Sample['Condition'].apply(lambda x: len(x) <= 20)]

MergedData_Sample['Condition'].unique()
```

<h4 style='color:skyblue;'>Screen Size</h4>

```python
MergedData_Sample['Screen Size'].unique()
```

```python
# Standardize the values in the 'Screen Size' column and convert to floats
MergedData_Sample['Screen Size'] = MergedData_Sample['Screen Size'].str.extract(r'(\d+\.?\d*)').astype(float)

# Check the new unique values
MergedData_Sample['Screen Size'].unique()
```

<h4 style='color:skyblue;'>Storage Type</h4>

```python
MergedData_Sample['Storage Type'].unique()
```

```python
# Standardize the storage type values
storage_type_mapping = {
    'SSD (Solid State Drive)': 'SSD',
    'HDD (Hard Disk Drive)': 'HDD',
    'Hybrid (HDD/SSD)': 'Hybrid',
    'HDD + SSD': 'Hybrid',
    'NVMe SSD': 'SSD',
    'NVMe': 'SSD',
    'Touchscreen': 'Unknown',
    'HDD/SSD': 'Hybrid',
    '256 G': 'Unknown',
    'Solid State Drive (SSD)': 'SSD',
    'NVMe (Solid State Drive)': 'SSD',
    '.': 'Unknown',
    'HDD / SSD': 'Hybrid',
    'NVMe m.2 SSD': 'SSD',
    'na': 'Unknown',
    'SSD + eMMC': 'Hybrid',
    'eMMC Solid State SSD': 'eMMC',
    'M.2 SOLID STATE': 'SSD',
    'NVMe (Non-Volatile Memory Express)': 'SSD',
    'SSD + HDD': 'Hybrid',
    'NVMe m.2': 'SSD'
}

# Replace using the mapping dictionary
MergedData_Sample['Storage Type'] = MergedData_Sample['Storage Type'].replace(storage_type_mapping)

# Check the new unique values
MergedData_Sample['Storage Type'].unique()
```

<h4 style='color:skyblue;'>Price</h4>

```python
MergedData_Sample['Price'].unique()
```

```python
# Use only the first price if there are multiple prices
MergedData_Sample['Price'] = MergedData_Sample['Price'].apply(lambda x: x.split(' to ')[0] if ' to ' in x else x)

# Remove all non-float characters from 'Price' and convert the column to a float data type
MergedData_Sample['Price'] = MergedData_Sample['Price'].replace('[\$,]', '', regex=True).astype(float)

# Check the new unique values
MergedData_Sample['Price'].unique()
```

<strong><h2 style='color:skyblue;'>Grouping & Aggregating </h2><strong>

```python
# Create object with brands grouped
brand_groups = MergedData_Sample.groupby('Brand')
brand_groups
```

```python
# Some summary statistics for the price by brand
brand_groups['Price'].agg(['count', 'mean', 'max', 'min']).sort_values(by='mean', ascending=False).style.format({'mean': "{:.2f}", 'max': "{:.2f}", 'min': "{:.2f}"})
```

```python
# create aggregate summary statistics for the price by RAM size
ram_groups = MergedData_Sample.groupby('Ram Size')
ram_groups['Price'].agg(['count', 'mean', 'max', 'min']).sort_values(by='mean', ascending=False).style.format({'mean': "{:.2f}", 'max': "{:.2f}", 'min': "{:.2f}"})
```

```python
# Visualize the distribution of RAM sizes

import matplotlib.pyplot as plt

# Count the occurrences of each RAM size
ram_size_counts = MergedData_Sample['Ram Size'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
ram_size_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of RAM Sizes')
plt.xlabel('RAM Size')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
